```python
import numpy as np
import os,sys
import cv2
import glob
from matplotlib import pyplot as plt 
```

```python
#path where i want to create the plots folders
#after creating these folders, copy the graphs from each split to them
plotsIm_path = '/home/marmia/snic2020-6-41/Mariam/Mariam_Thesis/Notebooks_Master/trainings/trainings/train1'

subdirs = ['loss_split1', 'loss_split2','loss_split3']
for subdir in subdirs:
    if not os.path.exists(os.path.join(plotsIm_path, subdir)):
        os.makedirs(os.path.join(plotsIm_path, subdir))
    else:
        os.system('rm -rf %s/*' % os.path.join(plotsIm_path, subdir))
```

```python
split1 = sorted(glob.glob(plotsIm_path+'/'+subdirs[0]+'/*'))
split2 = sorted(glob.glob(plotsIm_path+'/'+subdirs[1]+'/*'))
split3 = sorted(glob.glob(plotsIm_path+'/'+subdirs[2]+'/*'))

for im1,im2,im3 in zip(split1,split2,split3):
    train1 = im1.split('train')[-1].replace('.png','')
    train2 = im2.split('train')[-1].replace('.png','')
    train3 = im3.split('train')[-1].replace('.png','')
    im_1 = cv2.imread(im1)
    im_1 = np.array(im_1)
    im_2 = cv2.imread(im2)
    im_2 = np.array(im_2)
    im_3 = cv2.imread(im3)
    im_3 = np.array(im_3)
    fig, ax = plt.subplots(1,3,figsize=(15,8))
    ax[0].imshow(im_1)
    ax[1].imshow(im_2)
    ax[2].imshow(im_3)
    ax[0].set_title('Train and validation loss curves-split1')
    ax[1].set_title('Train and validation loss curves-split2')
    ax[2].set_title('Train and validation loss curves-split3')
    ax[0].set_xlabel('Number of epochs')
    ax[1].set_xlabel('Number of epochs')
    ax[2].set_xlabel('Number of epochs')
    ax[0].set_ylabel('loss'+'-'+train1)
    ax[1].set_ylabel('loss'+'-'+train2)
    ax[2].set_ylabel('loss'+'-'+train3)
    plt.show()
```
