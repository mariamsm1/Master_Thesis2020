```python
#Corr01

import matplotlib.pyplot as plt
import pandas as pd
import numpy as np


df = pd.read_csv("/home/marmia/snic2020-6-41/Mariam/Mariam_Thesis/Results_Pipelines_Images/Results/Old_CP_Output/Old_Data/New_data12_C01/ClassifyObjects/Classify_Object_C01Nuclei_Corr01.csv") 

out = open('step_Area127_Grouped_5.csv', 'w')
print('Area', 'Step' , sep = ',', file = out)

#save images number in a dataframe to be able to access them
images = df.ImageNumber.unique()
for i in range(len(images)):
    step_list= []
    #locate the images in the file
    data = df.loc[df.ImageNumber == images[i]].reset_index()
    #locate intensities for each image
    area = data.AreaShape_Area
    for j in range(len(area)):
        #empty the step for each intensty value
        step = 0
        while True:
            if data.AreaShape_Area[j] < step:
                step_list.append(step)
                break
            step = step+5
    for x in step_list:
        print(images[i],x, sep =',', file = out)
            
        
df = pd.read_csv("step_Area127_Grouped_5.csv") 

step = df.Step
plt.rcParams["figure.figsize"] = (45,8)
plt.hist(step, bins = len(np.arange(0,7000,100)))
plt.xticks(np.arange(7000, step=100))
plt.xlabel('Nuclei Area')
plt.ylabel('Number of Nuclei')
plt.savefig('d0_Area127_images_step5.png')
```
